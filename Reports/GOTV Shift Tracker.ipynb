```python
(select
	t2.fo_name,
	t1.event_date,
    t1.location,
    t1.role,
    t1.event_shift_id,
 	t3.current_shift_status_name,
 	case
 		when t3.current_shift_status_name in ('Left Msg', 'Confirmed', 'Conf Twice') then 'Confirmed'
 		when t3.current_shift_status_name in ('Scheduled', 'Sched-Web','Confrm-Web') then 'Unconfirmed' else 'Other' 
 	end as status_bucket,
 	sum(case when t3.current_shift_status_name in ('Left Msg', 'Confirmed', 'Conf Twice') then 1 else 0 end) as shifts_confirmed,
 	sum(case when t3.current_shift_status_name in ('Scheduled', 'Sched-Web','Confrm-Web') then 1 else 0 end) as shifts_unconfirmed,
    count(t2.myc_van_id) as cnt,
     'PTG Data Last Updated: '||to_char(convert_timezone('America/Chicago',getdate()),'Mon DD, YYYY HH12:MI am') as report_time
from
(select 
	case
    	when event_id = 406034 then 'Dry Run 1 10/10/20'
        when event_id = 406035 then 'Dry Run 1 10/11/20'
        when event_id = 406042 then 'Dry Run 2 10/17/20'
        when event_id = 406046 then 'Dry Run 2 10/18/20'
        when event_id = 406036 then 'GOTEV 10/21/20'
        when event_id = 406037 then 'GOTEV 10/22/20'
        when event_id = 406038 then 'GOTEV 10/23/20'
        when event_id = 406039 then 'GOTEV 10/24/20'
        when event_id = 406041 then 'GOTEV 10/25/20'
        when event_id = 406047 then 'GOTEV 10/26/20'
        when event_id = 406040 then 'GOTEV 10/27/20'
        when event_id = 406048 then 'GOTEV 10/28/20'
        when event_id = 406049 then 'GOTEV 10/29/20'
        when event_id = 406050 then 'GOTV 10/31/20'
        when event_id = 406043 then 'GOTV 11/01/20'
        when event_id = 406044 then 'GOTV 11/02/20'
        when event_id = 406045 then 'GOTV 11/03/20'
    end as event_date,
 	case
 		when location_id = 36362 then 'GOTV in Madison, Warren County Area'
 		when location_id = 36363 then 'GOTV in Dallas County Area'
 		when location_id = 36364 then 'GOTV in Sioux City Area'
 		when location_id = 36365 then 'GOTV in Spencer Area'
 		when location_id = 36366 then 'GOTV in Ankeny Area'
 		when location_id = 36367 then 'GOTV in Grimes Area'
 		when location_id = 36368 then 'GOTV in Beaverdale Area'
 		when location_id = 36369 then 'GOTV in ISU Area'
 		when location_id = 36370 then 'GOTV in Cerro Gordo, Cicksaw, Floyd Area'
 		when location_id = 36371 then 'GOTV in Northeast Iowa'
 		when location_id = 36372 then 'GOTV in Manchester Area'
 		when location_id = 36373 then 'GOTV in Waterloo/Waverly Area'
 		when location_id = 36374 then 'GOTV in Cedar Rapids Area'
 		when location_id = 36375 then 'GOTV in Grinnel Area'
 		when location_id = 36376 then 'GOTV in Johnson and Cedar County Area'
 		when location_id = 36377 then 'GOTV in Iowa City and Southcentral Iowa'
 		when location_id = 36378 then 'GOTV in Iowa City and Southeast Iowa'
 		when location_id = 36379 then 'GOTV in Scott County and Southeast Iowa'
 		when location_id = 36380 then 'GOTV in Muscatine and Scott County Area'
 		when location_id = 36402 then 'GOTV in OOS Phonebank'
 		when location_id = 36381 then 'GOTV in Southwest Iowa'
 		when location_id = 36382 then 'GOTV in Northwest Iowa'
 		when location_id = 36383 then 'GOTV in Polk County'
 		when location_id = 36384 then 'GOTV in North Central Iowa'
 		when location_id = 36385 then 'GOTV in Northeast Iowa'
 		when location_id = 36386 then 'GOTV in Linn Area'
 		when location_id = 36387 then 'GOTV in Johnson Area'
 		when location_id = 36388 then 'GOTV in Southeast Iowa'
 		when location_id = 36403 then 'GOTV in OOS Dialer'
 	end as location,
    case 
    	when event_role_id = 337854 and location_id in ('36362', '36363', '36364', '36365', '36366', '36367', '36368', '36369', '36370', '36371', '36372', '36373', '36374', '36375', '36376', '36377', '36378', '36379', '36380', '36402') then 'VPB Phonebanker'
        when event_role_id = 361667 and location_id in ('36381', '36382', '36383', '36384', '36385', '36386', '36387', '36388', '36403') then 'Dialer'
 		else 'Error'
	end as role,
    event_shift_id,
 	event_signup_id,
    myc_van_id
from my_state_van.event_signups
where event_role_id in (337854, 361667)
	and event_id is not null and event_id in ('406034', '406035', '406042', '406046', '406036', '406037', '406038', '406039', '406041', '406047', '406040', '406048', '406049', '406050', '406043', '406044', '406045')) t1 
left join
my_state_van.coord20_myc_005_person t2 on t1.myc_van_id = t2.myc_van_id
left join 
my_state_van.coord20_myc_003_event_shifts t3 on t1.event_signup_id = t3.event_signup_id
where t3.current_shift_status_name in ('Scheduled', 'Sched-Web', 'Confrm-Web', 'Conf Twice', 'Left Msg', 'Confirmed', 'Completed')
group by 1, 2, 3, 4, 5, 6
order by 1)
```
